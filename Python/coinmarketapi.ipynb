```python
import pandas as pd
import numpy as np
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json, os
```

```python
pd.set_option('display.max_columns', None)
```

```python
def api_runner():
        global file
        url = 'https://sandbox-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
        parameters = {
        'start':'1',
        'limit':'5000',
        'convert':'USD'
        }
        headers = {
        'Accepts': 'application/json',
        'X-CMC_PRO_API_KEY': '93c13268-7c12-43ce-8c8e-b3ba7ff26817',
        }

        session = Session()
        session.headers.update(headers)

        try:
            response = session.get(url, params=parameters)
            data = json.loads(response.text)
            #print(data)
        except (ConnectionError, Timeout, TooManyRedirects) as e:
            print(e)

        file2 = pd.json_normalize(data['data'])
        file2['timestamp']=pd.to_datetime('now')
        file = file.append(file2)

        if not os.path.isfile(r'/home/awal/Documents/xl/api.csv'):
            file.to_csv(r'/home/awal/Documents/xl/api.csv', header='column_names')
        else:
            file.to_csv(r'/home/awal/Documents/xl/api.csv', mode='a', header=False)
```

```python
import os
from time import time, sleep

for i in range(333):
    api_runner()
    print('API Runner completed Successfully!!')
    sleep(60) #sleep for every minute
    print("\n")
exit()
```
